```python
import requests
import os
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.environ.get("API_KEY")
```

```python
city = "des-moines"
state = "ia"
country = "us"
url = f"http://api.openweathermap.org/geo/1.0/direct?q={city},{state},{country}&appid={API_KEY}"
geocoding_res = requests.get(url)
geocoding_res.json()
```

```python
lat = geocoding_res.json()[0]["lat"]
lon = geocoding_res.json()[0]["lon"]
url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API_KEY}"
current_weather_res = requests.get(url)
current_weather_res.json()
```

```python
url = f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={API_KEY}"
forecast_res = requests.get(url)
forecast_res.json()
```

```python
url = f"http://api.openweathermap.org/data/2.5/air_pollution/forecast?lat={lat}&lon={lon}&appid={API_KEY}"
one_call_res = requests.get(url)
one_call_res.json()
```
